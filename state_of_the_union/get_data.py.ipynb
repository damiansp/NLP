Wikisource has all of the State of the Union Addresses available from <a href=""Wikisource has all of the State of the Union Addresses available from <a href="https://en.wikisource.org/wiki/Portal:State_of_the_Union_Speeches_by_United_States_Presidents">this landing page</a>. Get and save the speeches.

```python
from   bs4 import BeautifulSoup
from   requests_html import HTMLSession
```

```python
DATA = '../data'
DOMAIN = 'https://en.wikisource.org'
LANDING_PAGE = (
    DOMAIN 
    + '/wiki/Portal:'
    'State_of_the_Union_Speeches_by_United_States_Presidents')
SESSION = HTMLSession()
```

```python
def get_soup(url):
    res = SESSION.get(url)
    html = res.html.html
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
def get_links(soup):
    links = soup.find_all(href=True)
    links = [DOMAIN + link['href'] for link in links 
             if 'State_of_the_Union_Address' in link['href']]
    return links
```

```python
def get_speech_text(url):
    soup = get_soup(url)
    ps = soup.find_all('p')[1:-1]
    text = [p.text for p in ps]
    return '\n'.join(text)
```

```python
soup = get_soup(LANDING_PAGE)
links = get_links(soup)
for link in links:
    try:
        title = (link.split('/')[-1]
                 .replace('_State_of_the_Union_Address', '')
                 .replace('%27', "'"))
        print(title)
        text = get_speech_text(link)
        with open(f'{DATA}/{title}.txt', 'w') as f:
            f.write(text)
    except BaseException as e:
        print()
        print(link)
        print(e)
```
